```python
book_number = '17814'

gutenberg_root = '/data/projects/catss/sources/gutenberg/el/aleph.gutenberg.org'
output_root = './sources/gutenberg'

try:
    book_dir = os.path.join(output_root, book_number)
    os.mkdir(book_dir)
except FileExistsError:
    pass

gutenberg_path = os.path.join(gutenberg_root, *book_number[:-1], book_number, '*.zip')
for filename in glob(gutenberg_path):
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall(book_dir)
        
    
    
```
